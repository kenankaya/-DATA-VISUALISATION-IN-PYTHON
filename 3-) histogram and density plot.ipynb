```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds=sns.load_dataset('diamonds')
df=diamonds.copy()
categories_of_cut=["Fair","Good","Very Good","Premium","Ideal"]
df.cut=df.cut.astype(CategoricalDtype(categories=categories_of_cut,ordered=True))
df.head()
```

```python
#distplot===>>> distribution plot=====>>> gives information regarding the distiribution of numerical variables
sns.distplot(df["price"],kde=False);
```

```python
sns.distplot(df["price"],bins=10,kde=False);
```
